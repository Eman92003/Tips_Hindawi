```python
from urllib.parse import urlparse, parse_qs
from youtube_transcript_api import YouTubeTranscriptApi


def extract_video_id(url: str) -> str:
    """
    Extract the YouTube video ID from a URL.
    Raises ValueError if no 'v' parameter is found.
    """
    parsed = urlparse(url)
    qs = parse_qs(parsed.query)
    video_ids = qs.get('v')

    if not video_ids:
        raise ValueError(f"No video id found in URL: {url}")

    return video_ids[0]


if __name__ == "__main__":
    url = "https://www.youtube.com/watch?v=SkidyDQuupA"

    video_id = extract_video_id(url)

    api = YouTubeTranscriptApi()
    fetched = api.fetch(video_id, languages=['en'])  # returns a FetchedTranscript

    text = "\n".join(snippet.text for snippet in fetched)
    print(text)
```

```python
def simple_clean(text):
    lines = text.split("\n")                 # تقسيم النص لأسطر
    lines = [line.strip() for line in lines] # إزالة الفراغات من أول ونهاية السطر
    lines = [line for line in lines if line] # إزالة الأسطر الفارغة

    skip_words = ["pause the video", "thanks for watching", "feel free to"]
    cleaned_lines = []
    for line in lines:
        if not any(word.lower() in line.lower() for word in skip_words):
            cleaned_lines.append(line)
    
    cleaned_text = " ".join(cleaned_lines)
    return cleaned_text

cleaned_text = simple_clean(text)
print(cleaned_text)  
```

## **Use Bart-large-CNN to summarize**

```python
from transformers import pipeline

summarize = pipeline("summarization", model="facebook/bart-large-cnn")      
```

```python
print(summarize(cleaned_text, max_length=130, min_length=30, do_sample=False))
```

**Limit input size of bart-large-cnn made cuda error**  <br>
**splitting text into chunks that are in the range of the model**

```python
chunks = [cleaned_text[i:i+1000] for i in range(0, len(cleaned_text), 1000)]
summaries = [summarize(chunk, max_length=100, min_length=30, do_sample=False) for chunk in chunks]
final_summary = " ".join([s[0]['summary_text'] for s in summaries])
print(final_summary)
```

```python
final_output = summarize(final_summary, max_length=130, min_length=30, do_sample=False)
```
